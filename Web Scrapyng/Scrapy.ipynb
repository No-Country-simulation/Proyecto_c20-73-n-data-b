```python
# Ran this on Google Colab
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
from io import StringIO

from random import randint

# list of countries of major countries (most minor countries won't have even 1 influencer on some of the platforms)
countries = [
    'All Countries', 'Afghanistan','Argentina', 'Australia', 'Austria', 'Belgium', 'Bolivia', 'Brazil', 'Canada', 'Chile', 'Colombia',
    'Costa Rica', 'Czechia', 'Denmark', 'Dominican Republic', 'Ecuador', 'Egypt', 'El Salvador', 'Finland',
    'France', 'Germany', 'Greece', 'Guatemala', 'Hong Kong', 'India', 'Indonesia', 'Italy', 'Japan', 'Kuwait',
    'Lebanon', 'Malaysia', 'Mexico', 'Morocco', 'Netherlands', 'Norway', 'Pakistan', 'Panama', 'Paraguay',
    'Peru', 'Philippines', 'Poland', 'Portugal', 'Puerto Rico', 'Romania', 'Russia', 'Saudi Arabia', 'Singapore',
    'Slovakia', 'South Africa', 'South Korea', 'Spain', 'Sweden', 'Switzerland', 'Thailand', 'Turkey', 'Ukraine',
    'United Arab Emirates', 'United Kingdom', 'United States', 'Uruguay'
]

# convert each country name to lowercase and replace spaces with dashes as urls are in this format
formatted_countries = [country.lower().replace(' ', '-') for country in countries]

# list of available social media platforms on website
social_medias = ['instagram', 'youtube', 'tiktok', 'threads']

# Set headers with a User-Agent to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

def scrape_webpage(url, headers, country, social):
    try:
        response = requests.get(url, headers=headers)

        # check if response status code is 200 
        if response.status_code == 200:
            print(f"{country} Scraping Started!")

            # read HTML content and convert it into DataFrame
            df = pd.read_html(StringIO(response.text))
            data = pd.DataFrame(df[0])

            # drop irrelavent columns
            try:
                data.drop(columns=['SAVE', 'INVITE TO CAMPAIGN'], inplace=True)
            except:
                print(f"Data of {country} of {social}")

            print(data.shape)
            print(f"{country} Scraping successful!")

            return data

        else:
            print("Failed to scrape:", response.status_code)
            # Handle other response status codes

    except requests.exceptions.RequestException as e:
        print("Error:", e)

# introduce random delays between requests
def delay_request():
    time.sleep(randint(1, 3))  # Random delay between 1 to 3 seconds

url = "https://starngage.com/plus/en-us/influencer/ranking"

itr = 0

# I'm just scraping overall data in this case for all countries comment this
formatted_countries = ['all-countries']

# scraping loop for each country and social media
for country in formatted_countries:
    folder_name = country
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    for social in social_medias:
        url_new = url + "/" + social + "/" + country

        csv_file_path = os.path.join(folder_name, f'{social}_data_{country}.csv')

        # scrape webpage and save data to CSV
        df = scrape_webpage(url_new, headers, country, social)
        df.to_csv(csv_file_path, index=False)

        # add random delay between requests to prevent browser to block us
        delay_request()
```
